```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define tickers on Yahoo Finance
tickers = ["BTC-USD", "ETH-USD", "SOL-USD", "XRP-USD", "ADA-USD"]

# Fetch historical prices
data = yf.download(tickers, start="2023-01-01", end="2025-01-31")

print("Downloaded Data Sample:")
print(data.head())
print(data.tail())

# Check if "Adj Close" exists, otherwise use "Close"
if "Adj Close" in data.columns:
    data = data["Adj Close"]
elif "Close" in data.columns:
    data = data["Close"]
else:
    raise KeyError("Neither 'Adj Close' nor 'Close' column found in the downloaded data.")

# Compute the correlation matrix
correlation_matrix = data.corr()

print("Correlation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of BTC, ETH, XRP, ADA and SOL (Daily Timeframe)")
plt.show()
```
